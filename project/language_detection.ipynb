```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv("language.csv")  # Load your data file

# Process data
if data.isnull().sum().sum() > 0:
    data = data.dropna()  # Remove missing values

x = np.array(data['Text'])  # Feature (text) column
y = np.array(data['language'])  # Target (language labels)

cv = CountVectorizer()
X = cv.fit_transform(x)  # Converting text into numerical data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Train the language detection model
model = MultinomialNB()
model.fit(X_train, y_train)

def detect_language(text):
    """
    Detects the language of the given text.
    """
    input_data = cv.transform([text]).toarray()
    detected_language = model.predict(input_data)[0]
    return detected_language
```
